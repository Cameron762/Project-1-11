```python
import pandas as pd
from pathlib import Path
```

```python
df = Path("../Resources/data.csv")
dfdf= pd.read_csv(df)
dfdf.head(10)
```

```python
cleandf = dfdf.sort_values(['Country Name'],ascending = False)
cleandf.head()
```

```python
df_clean = cleandf.rename(columns={"GDP (current US$) [NY.GDP.MKTP.CD]":"GDP in USD",
"International tourism, receipts (current US$) [ST.INT.RCPT.CD]":"International Tourism Receipts USD",
"International tourism, expenditures (current US$) [ST.INT.XPND.CD]":"International Tourism Expenditures USD",
"International tourism, number of arrivals [ST.INT.ARVL]":"International Tourism Arrivals"})
df_clean.head()
```

```python
dfclean = df_clean.drop(["Time Code","Country Code",],axis = 1)
dfclean.head()
```

```python
dfclean.count()
```

```python
clean = dfclean.dropna(how = "any")
clean.count()
```

```python
cleanddddf = clean.drop(clean[clean['International Tourism Receipts USD'] == ".."].index, inplace = False)
cleanddddf.count()
```

```python
cleanddddf.to_csv("../Resources/cleandata.csv")
file = Path("../Resources/cleandata.csv")
pd.read_csv(file)
```
