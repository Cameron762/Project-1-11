```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
df = Path("../Resources/tourismdata.csv")
dfdf= pd.read_csv(df)
dfdf.head(10)
```

```python
dfdf.dtypes
```

```python
cleandf = dfdf.sort_values(['Country Name'],ascending = False)
cleandf.head()
```

```python
renamed_df = cleandf.rename(columns={"Time":"Year", 
                                "International tourism, expenditures (current US$) [ST.INT.XPND.CD]":"International Tourism Expenditures (US$)", 
                                "International tourism, receipts (current US$) [ST.INT.RCPT.CD]":"International Tourism Receipts (US$)", 
                                "International tourism, number of arrivals [ST.INT.ARVL]":"# of arrivals", 
                                "International tourism, number of departures [ST.INT.DPRT]":"# of departures", 
                                "Population, total [SP.POP.TOTL]":"Total Population",
                                  "GDP (current US$) [NY.GDP.MKTP.CD]":"GDP (US$)"})
renamed_df.head()
```

```python
dfclean = renamed_df.drop(["Time Code","Country Code",],axis = 1)
dfclean.head()
```

```python
dfclean.count()
```

```python
dfclean
```

```python
cleanrep = dfclean.replace('..', np.nan)
clean = cleanrep.dropna(how = "any")

clean.count()
```
