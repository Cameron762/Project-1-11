```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
df = Path("../Resources/tourismdata.csv")
dfdf= pd.read_csv(df)
dfdf.head(10)
```

```python
dfdf.dtypes
```

```python
cleandf = dfdf.sort_values(['Country Name'],ascending = True)
cleandf.head()
```

```python
renamed_df = cleandf.rename(columns={"Time":"Year", 
                                "International tourism, expenditures (current US$) [ST.INT.XPND.CD]":"International Tourism Expenditures (US$)", 
                                "International tourism, receipts (current US$) [ST.INT.RCPT.CD]":"International Tourism Receipts (US$)", 
                                "International tourism, number of arrivals [ST.INT.ARVL]":"# of arrivals", 
                                "International tourism, number of departures [ST.INT.DPRT]":"# of departures", 
                                "Population, total [SP.POP.TOTL]":"Total Population",
                                  "GDP (current US$) [NY.GDP.MKTP.CD]":"GDP (US$)"})
renamed_df.head()
```

```python
dfclean = renamed_df.drop(["Time Code","Country Code","# of departures"],axis = 1)
dfclean.head()
```

```python
dfclean.count()
```

```python
cleanrep = dfclean.replace('..', np.nan)
clean = cleanrep.dropna(how = "any")
clean.count()
```

```python
clean = clean.copy()
clean['GDP (US$)'] = clean['GDP (US$)'].astype(int)
clean['# of arrivals'] = clean['# of arrivals'].astype(int)
clean['International Tourism Receipts (US$)'] = clean['International Tourism Receipts (US$)'].astype(int)
clean['International Tourism Expenditures (US$)'] = clean['International Tourism Expenditures (US$)'].astype(int)
clean['Total Population'] = clean['Total Population'].astype(int)
```

```python
clean.dtypes
```

```python
cleanround1 = clean.round({'Total Population': 0})
cleanround2 = cleanround1.round({'# of arrivals':0})
cleanround3 = cleanround2.round({'GDP (US$)':0})
cleanround3
```

```python
cleanround1.dtypes
```

```python
clean.to_csv("../Resources/clean3data.csv")
file = Path("../Resources/clean3data.csv")
pd.read_csv(file)
```
