```python
import pandas as pd
from pathlib import Path
```

```python
country = Path("Resources/countrypgdp.csv")
countrydf = pd.read_csv(country, encoding = "utf-8")
countrydf.head()
```

```python
c_country = countrydf.drop(["GDP"],axis = 1)
clean_country = c_country.sort_values(["Country Name","Year"],ascending=True)
clean_country.head()
```

```python
countryGDP = clean_country.drop(clean_country[clean_country['Year'] == 2021].index, inplace = False)
countryGDP.head(20)
```

```python
countryGDP.to_csv(("Resources/countryGDP.csv"))
```

```python
reciepts = Path("../Resources/tourism-recipts.csv")
recieptsdf = pd.read_csv(reciepts, encoding = "utf-8")
recieptsdf.head()
```

```python
creciepts = recieptsdf[(recieptsdf["year"]>= 2017) & (recieptsdf["year"] <= 2020)]
creciepts.head()
```

```python
c_reciepts = creciepts.rename(columns={"name":"Country","year":"Year","value_$":"Total $ of Tourism Reciepts"})
clean_reciepts = c_reciepts.drop(["code"],axis = 1)
clean_reciepts.head(20)
```

```python
clean_reciepts.to_csv("../Resources/reciepts.csv")
```
